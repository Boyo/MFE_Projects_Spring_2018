```python
data_pre = data[data['Year'] < 1945]
data_post=data[data['Year'] >= 1945]
# T test
for i in range(3,14):
    mean1=data_pre.iloc[:,i].mean(0) # mean (by column) 
    stdev1=data_pre.iloc[:,i].std(0)
    mean2=data_post.iloc[:,i].mean(0) # mean (by column) 
    stdev2=data_post.iloc[:,i].std(0)
    s=math.sqrt(stdev1/len(data_pre.iloc[:,i])+stdev2/len(data_post.iloc[:,i]))
    t=(mean1-mean2)/s
    degree_freedom_numerator=np.power((stdev1/len(data_pre.iloc[:,i])+stdev2/len(data_post.iloc[:,i])),2)
    degree_freedom_denominator=np.power((stdev1/len(data_pre.iloc[:,i])),2)/(len(data_pre.iloc[:,i])-1)+np.power(stdev2/len(data_post.iloc[:,i]),2)/(len(data_post.iloc[:,i])-1)
    degree_freedom=degree_freedom_numerator/degree_freedom_denominator
    s_distribution = np.random.standard_t(degree_freedom, size=100000)
    p_value=(np.sum(s_distribution<-abs(t))+np.sum(s_distribution>abs(t)))/float(len(s_distribution))
    print(i-2,p_value)
```

# Efficient Markets

```python
n=len(data.iloc[:,1])
Result_beta=[]
for i in range(3,14):
    Y=data.iloc[1:n,i]
    X=data.iloc[0:n-1,i]
    one= pd.Series(1, index=range(0,n-1))
    X=pd.concat([one, X], axis=1)
    Beta=np.dot(inv(np.dot(X.T,X)),np.dot(X.T,Y))
    Result_beta=Result_beta+[Beta]
    print("c: ", Beta[0],"phi: ",Beta[1])
```

```python
# Ljung-Box test
from scipy.stats import chi2
lag=1000 #approximately 4 years (mettre à 1000)
quantile=0.05
for i in range(3,14):
    s=0
    for h in range(1,lag+1):
        sample_autocorrelation=data_pre.iloc[:,i].autocorr(h) # sample autocorrelation (by column) 
        s=s+np.power(sample_autocorrelation,2)/(n-h)
    Q=n*(n+2)*s
    s_distribution = chi2.isf(1-quantile, lag)
    if (Q>s_distribution):
        decision=" => reject"
    else:
        decision=" => accept"
    print(i-2,Q,s_distribution,decision)
```

```python
# T Test
for i in range(0,10):
    Y=data.iloc[1:n,i+3]
    X=data.iloc[0:n-1,i+3]
    c=Result_beta[i][0]
    beta=Result_beta[i][1]
    ss_tot=0
    ss_res=0
    for k in range(0,n-1):
        ss_tot=ss_tot+np.power((Y.iloc[k]-Y.mean(0)),2)
        ss_res=ss_res+np.power(Y.iloc[k]-(c+X.iloc[k]*beta),2)/n
    sb=math.sqrt(ss_res/ss_tot)
    t=beta/sb
    print(t)
    s_distribution = np.random.standard_t(n, size=100000)
    p_value=(np.sum(s_distribution<-abs(t))+np.sum(s_distribution>abs(t)))/float(len(s_distribution))
    print(i, p_value)
```

# Data-Mining

```python
n=len(data.iloc[:,1])
for i in range(3,6): # first 3 deciles
    # (a)
    Y=data.iloc[1:n,13]
    X=data.iloc[0:n-1,i]
    one= pd.Series(1, index=range(0,n-1))
    X2=pd.concat([one, X], axis=1)
    Beta=np.dot(inv(np.dot(X2.T,X2)),np.dot(X2.T,Y))
    print("c: ", Beta[0],"phi: ",Beta[1])
    
    # (b)
    # T Test
    c=Beta[0]
    beta=Beta[1]
    ss_tot=0
    ss_res=0
    for k in range(0,n-1):
        ss_tot=ss_tot+np.power((Y.iloc[k]-Y.mean(0)),2)
        ss_res=ss_res+np.power(Y.iloc[k]-(c+X.iloc[k]*beta),2)/n
    sb=math.sqrt(ss_res/ss_tot)
    t=beta/sb
    print("t-value: ", t)
    s_distribution = np.random.standard_t(n, size=100000)
    p_value=(np.sum(s_distribution<-abs(t))+np.sum(s_distribution>abs(t)))/float(len(s_distribution))
    print(i, p_value)
```
